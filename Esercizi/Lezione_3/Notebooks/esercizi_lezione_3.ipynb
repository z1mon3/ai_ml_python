# **Descrizione dei dati**

**Dataset:**
- Alberghi: arrivi, presenze
- Agriturismi: arrivi, presenze
- Villaggi e Campeggi: arrivi, presenze

**fonte:** http://www.datiopen.it/it





***Definizione di arrivi turistici:*** 


 *il numero di clienti, italiani e stranieri, ospitati negli esercizi ricettivi (alberghieri o complementari) nel periodo considerato.*



***Definizione di presenze turistiche:***

 Il numero delle notti trascorse dai clienti negli esercizi ricettivi (alberghieri o complementari). 

**Esempio di lettua dei dati:**

In **Piemonte**	nel **2008**	si sono verificati **2747466 arrivi**   per un totale di **7499879 notti**	trascrorse.
