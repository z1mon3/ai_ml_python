**Download e modifica di file csv**

```python
def test_dataset_anno_2017():
    path = "http://www.comune.torino.it/opendata/statistica/nascite_decessi/nati_circ_17.csv"

    # Indicazioni nella lettura del file csv
    # il file contiene caratteri non compatibili con la codifica utf-8
    # possibili codifiche: 
    #       - latin1 = utile per la lingua italiana (es. lettere accentate)
    #       - ISO-8859-1 = utile come estensione generica a utf-8
    df = pd.read_csv(path, encoding="latin1")

    # Questa opzione risulta estremamente utile nel caso in cui 
    # i nomi delle colonee presentino "white space".
    # Ad esempio "CIRCOSCRIZIONE ", " CIRCOSCRIZIONE" e "CIRCOSCRIZIONE"
    # sono tre parloe differenti per via degli spazi prima e dopo i caratteri.
    # In questo modo si è sempre certi di avere stringhe ben compattate.
    df = df.rename(str.strip, axis='columns')

    # "Nome della colonna" in DataFrame.columns, testa se una colonna esiste 
    # o meno in un dataframe
    c = "CIRCOSCRIZIONE" in df.columns
    f = "N_FEMMINE" in df.columns
    m = "N_MASCHI" in df.columns
    return (c, f, m)
```

```python
def test_dataset_anno_2018():

    # In questo caso il file csv non usa il separatore standard (la virgola ",") ma il ";".
    df = pd.read_csv("http://www.comune.torino.it/statistica/dati/2018/csv/D42018.csv", sep=";")
    df = df.rename(str.strip, axis='columns')

    c = "Circoscrizione di residenza" in df.columns
    f = "Femmine" in df.columns
    m = "Maschi" in df.columns
    return (c, f, m)

```

Unione di due DataFrame

```python
import pandas as pd

df1 = pd.DataFrame({"Col1": [1, 2, 3], 
                    "Col2": ["A", "B", "C"],
                    "Col3": ["Ford", "Fiat", "Opel"]})


df2 = pd.DataFrame({"C1": [4, 5, 6],
                    "C2": ["D", "E", "F"]})


# aggiungere una colonVolumena di un DataFrame ad un altro
df_j = df2.join(df1['Col3'])
print("joint")
display(df_j)
print("\n")

# affiancare due dataframe
df_c = pd.concat([df1, df2], axis=1)
print("concat")
display(df_c)
print("\n")

```
